```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Arun", "Priya", "Kumar", "Divya", "Rahul"],
    "Maths": [90, 85, 70, 95, 80],
    "Science": [88, 90, 65, 92, 75],
    "English": [85, 80, 75, 90, 82]
}

df = pd.DataFrame(data)
df
```

```python
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df
```

```python
df["Average"] = round(df["Total"] / 3, 2)
df
```

```python
df["Result"] = ["Pass" if x >= 35 else "Fail" for x in df["Average"]]

df
```

```python
topper = df.loc[df["Total"].idxmax()]
print(topper)
```

```python
print("Class Average =", round(df["Average"].mean(), 2))
print("Highest Mark =", df["Total"].max())
print("Lowest Mark =", df["Total"].min())
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Total"])
plt.title("Student Performance Analysis")
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.show()
```

```python
df.to_csv("student_report.csv", index=False)
print("Report Generated Successfully")
```
